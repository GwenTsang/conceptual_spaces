```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def plot_cone(radius, height):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # Create data for the circumference of the base
    u = np.linspace(0, 2 * np.pi, 200)
    x_base = radius * np.cos(u)
    y_base = radius * np.sin(u)

    # Plot the circumference of the base
    ax.plot(x_base, y_base, zs=0, zdir='z', color='black')

    # Calculate the length of the segments
    segment_length = np.sqrt(radius**2 + height**2)

    # Plot the two segments from top to base (reversed direction)
    ax.plot([0, radius], [0, 0], zs=[height, 0], zdir='z', color='black')
    ax.plot([0, -radius], [0, 0], zs=[height, 0], zdir='z', color='black')
    ax.plot([radius, 0], [0, 0], zs=[0, 0], zdir='z', color='black')
    ax.plot([-radius, 0], [0, 0], zs=[0, 0], zdir='z', color='black')

    # Calculate the coordinates for the base of the second triangle
    base_x = radius * np.cos(np.pi/2)
    base_y = radius * np.sin(np.pi/2)

    # Plot the base of the second triangle perpendicular to the first one
    ax.plot([base_x, -base_x], [base_y, -base_y], zs=[0, 0], zdir='z', color='black')

    # Plot the segments of the second triangle
    ax.plot([base_x, 0], [base_y, 0], zs=[0, height], zdir='z', color='black')
    ax.plot([-base_x, 0], [-base_y, 0], zs=[0, height], zdir='z', color='black')

    # Duplicate the triangles on the z-axis
    ax.plot([0, radius], [0, 0], zs=[-height, 0], zdir='z', color='black')
    ax.plot([0, -radius], [0, 0], zs=[-height, 0], zdir='z', color='black')
    ax.plot([base_x, 0], [base_y, 0], zs=[0, -height], zdir='z', color='black')
    ax.plot([-base_x, 0], [-base_y, 0], zs=[0, -height], zdir='z', color='black')

    # Color points with contours
    ax.scatter(0, 0, height, color='#FFFFFF', linewidths=1, edgecolors='black', s=40)  # Blanc White
    ax.scatter(0, 0, 0, color='#808080', linewidths=1, edgecolors='black', s=40)        # Gris Gray
    ax.scatter(0, 0, -height, color='#000000', linewidths=1, edgecolors='black', s=40)  # Noir Black

    ax.scatter(base_x, base_y, 0, color='#FFFF00', linewidths=1, edgecolors='black', s=40) #Bleu Blue
    ax.scatter(-base_x, -base_y, 0, color='#0000FF', linewidths=1, edgecolors='black', s=40) # Jaune Yellow

    ax.scatter(radius, 0, 0, color='#FF0080', linewidths=1, edgecolors='black', s=40)   # Magenta
    ax.scatter(-radius, 0, 0, color='#00FF60', linewidths=1, edgecolors='black', s=40)  # Vert-Turquoise

    # Remove axis markers
    ax.set_axis_off()

    # Set aspect ratio
    ax.set_box_aspect([1, 1, 1])

    ax.view_init(elev=16, azim=-100)

# Define radius and height variables
radius = 1
height = 2

# Plot the cone
plot_cone(radius, height)

plt.savefig("spindle.svg", format='svg')

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the 8 vertices of the cube
vertices = np.array([[0, 0, 0],  # white
                     [1, 0, 0],  # red
                     [1, 1, 0],  # yellow
                     [0, 1, 0],  # green
                     [0, 0, 1],  # blue
                     [1, 0, 1],  # magenta
                     [1, 1, 1],  # cyan
                     [0, 1, 1]]) # cyan

# Define the colors corresponding to each vertex
colors = np.array(['white', 'red', 'yellow', 'green', 'blue', 'magenta', 'black', 'cyan'])

# Define the 12 edges of the cube by connecting the vertices
edges = [(0, 1), (1, 2), (2, 3), (3, 0),
         (4, 5), (5, 6), (6, 7), (7, 4),
         (0, 4), (1, 5), (2, 6), (3, 7)]



# Plotting
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for edge in edges:
    v0, v1 = vertices[edge[0]], vertices[edge[1]]
    distance = np.linalg.norm(v1 - v0)
    ax.plot3D(*zip(*[v0, v1]), color='black', linewidth=2*distance)  # Adjust linewidth based on distance

# Plot the vertices with colors and black outline
for i in range(len(vertices)):
    ax.scatter(vertices[i][0], vertices[i][1], vertices[i][2], color=colors[i], s=80, edgecolors='black', linewidths=1)

# Add arrows for each axis
arrow_length = 1.2
inside_cube_offset = 0.1  # Offset to place the green arrow inside the cube

ax.quiver(1.2, 0, 0, arrow_length, 0, 0, color='red', arrow_length_ratio=0.1)
ax.quiver(0 + inside_cube_offset, 1.2, 0, 0, -arrow_length, 0, color='green', arrow_length_ratio=0.1)
ax.quiver(0, 0, 1.2, 0, 0, arrow_length, color='blue', arrow_length_ratio=0.1)



# Hide the axes
ax.set_axis_off()

# Save the plot as SVG
plt.savefig("cube_plot.svg", format='svg')

plt.show()
```
