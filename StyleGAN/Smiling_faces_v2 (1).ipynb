```python
#@title Isolate Smiling Faces and Save Latent Vectors
# Step 1: Install and import necessary libraries
!pip install -q datasets ftfy regex tqdm openai-clip Pillow huggingface_hub
!pip install --upgrade -q Pillow

import os
import zipfile
from tqdm.auto import tqdm
import torch
import clip
from PIL import Image
import numpy as np
import pandas as pd
from datasets import load_dataset
from huggingface_hub import hf_hub_download
from IPython.display import display

print("âœ… Step 1: Libraries installed and imported.")

# Step 2: Load metadata and manually download & extract images
print("\nðŸ”„ Step 2: Loading dataset components...")

try:
    # Part A: Load ONLY the metadata from the CSV file.
    print("   - Loading metadata from annotations.csv...")
    metadata_dataset = load_dataset("balgot/stylegan3-annotated", data_files="annotations.csv", split='train')
    print("   - Metadata loaded successfully.")

    # Part B: Manually download the images.zip file.
    print("   - Downloading images.zip from Hugging Face Hub...")
    zip_path = hf_hub_download(
        repo_id="balgot/stylegan3-annotated",
        filename="images.zip",
        repo_type="dataset"
    )
    print(f"   - Image archive downloaded to: {zip_path}")

    # Part C: Extract the zip file to a known directory.
    extract_path = "stylegan3_images"
    os.makedirs(extract_path, exist_ok=True)
    print(f"   - Unzipping images to '{extract_path}' directory...")
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        # Check if files are already extracted to avoid re-extracting
        # Using a sample file check
        sample_file_path = os.path.join(extract_path, "data", "stylegan_images", "00000.jpg")
        if not os.path.exists(sample_file_path):
            for member in tqdm(zip_ref.infolist(), desc="   - Extracting "):
                zip_ref.extract(member, extract_path)
        else:
            print("   - Images already extracted. Skipping extraction.")


except Exception as e:
    print(f"   - FATAL ERROR during data loading/extraction: {e}")
    exit()

print("âœ… Step 2: Dataset components are ready.")


# Step 3: Load the CLIP model
print("\nðŸ”„ Step 3: Loading CLIP model...")
device = "cuda" if torch.cuda.is_available() else "cpu"
try:
    model, preprocess = clip.load("ViT-L/14", device=device)
    print(f"   - Loaded CLIP ViT-L/14 on {device}.")
except Exception:
    print("   - Could not load ViT-L/14, falling back to ViT-B/32.")
    model, preprocess = clip.load("ViT-B/32", device=device)
    print(f"   - Loaded CLIP ViT-B/32 on {device}.")

# Define the attribute and create text prompts
attr_text = "smile"
text_prompts = [f"a photo of a face with no {attr_text}", f"a photo of a face with a {attr_text}"]
text_tokens = clip.tokenize(text_prompts).to(device)

print("âœ… Step 3: CLIP model loaded.")

# Step 4: Process images and filter for the "smile" attribute
print("\nðŸ”„ Step 4: Processing images to find smiling faces...")

smiling_faces_data = []

with torch.no_grad():
    for item in tqdm(metadata_dataset, desc="   - Classifying"):
        try:
            filename = f"{str(item['seed']).zfill(5)}.jpg"
            image_path = os.path.join(extract_path, "data", "stylegan_images", filename)

            pil_image = Image.open(image_path)
            image_tensor = preprocess(pil_image).unsqueeze(0).to(device)

            logits_per_image, _ = model(image_tensor, text_tokens)
            probs = logits_per_image.softmax(dim=-1).cpu().numpy()

            prediction = np.argmax(probs)
            confidence = np.max(probs)

            # If the model predicted "smile" (index 1), store all relevant data
            if prediction == 1:
                smiling_faces_data.append({
                    'full_path': image_path,
                    'seed': item['seed'],
                    'confidence': confidence,
                    'original_caption': item['text'],
                    'vector': item['vector']  # <-- CRITICAL: Save the latent vector
                })

        except FileNotFoundError:
            print(f"   - CRITICAL: Image file not found at {image_path}. Skipping.")
        except Exception as e:
            print(f"   - Warning: Could not process {image_path}. Error: {e}. Skipping.")

print("âœ… Step 4: Image processing complete.")


# Step 5: Display the results
print("\nðŸ“Š Step 5: Analyzing and Displaying Results")

if not smiling_faces_data:
    print("\nNo smiling faces were detected with the current settings. Exiting.")
    exit()

df_smiling = pd.DataFrame(smiling_faces_data)

# Fix the float32 dtype issue from the previous step to prevent errors
df_smiling['confidence'] = df_smiling['confidence'].astype('float64')

print(f"\nProcessed {len(metadata_dataset)} images.")
print(f"Found {len(df_smiling)} smiling faces.")

print("\nTop 5 results with highest confidence:")
display(df_smiling.sort_values(by='confidence', ascending=False).head())

print("\nðŸ–¼ï¸ Sample of detected smiling faces:")
num_samples_to_show = min(len(df_smiling), 6)
sample_df = df_smiling.nlargest(num_samples_to_show, 'confidence')

for index, row in sample_df.iterrows():
    print(f"--- Seed: {row['seed']}, Confidence: {row['confidence']:.4f} ---")
    img = Image.open(row['full_path'])
    img.thumbnail((256, 256))
    display(img)

print("âœ… Step 5: Analysis complete.")

# Step 6: Save Results (CSV and Vector Files)
print("\nðŸ’¾ Step 6: Saving results for future use...")

# Define output directories
output_dir = "smiling_faces_output"
vectors_dir = os.path.join(output_dir, "smiling_vectors")
os.makedirs(vectors_dir, exist_ok=True)

# --- 1. Save the comprehensive CSV file ---
csv_path = os.path.join(output_dir, "smiling_faces_data.csv")
print(f"   - Saving comprehensive data to: {csv_path}")
df_smiling.to_csv(csv_path, index=False)

# --- 2. Save each latent vector as a separate .npy file ---
print(f"   - Saving {len(df_smiling)} individual latent vectors to: {vectors_dir}")
print("     (This is the standard format for StyleGAN analysis tools)")

for index, row in tqdm(df_smiling.iterrows(), total=df_smiling.shape[0], desc="   - Saving vectors"):
    # Get the vector and seed from the DataFrame row
    vector_data = row['vector']
    seed = row['seed']

    # Define the output filename based on the seed, ensuring consistent naming
    vector_filename = f"{str(seed).zfill(5)}.npy"
    vector_output_path = os.path.join(vectors_dir, vector_filename)

    # Convert the vector list to a NumPy array with the correct data type
    # StyleGAN models typically use float32 for latent vectors
    np_vector = np.array(vector_data, dtype=np.float32)

    # Save the NumPy array to the .npy file
    np.save(vector_output_path, np_vector)

print("\nâœ… Step 6: All results saved successfully!")
print("You can now find your data in the 'smiling_faces_output' directory in the file browser.")
```

```python
# Step 6: Save Results (CSV and Vector Files)
print("\nðŸ’¾ Step 6: Saving results for future use...")

# Define output directories
output_dir = "smiling_faces_output"
vectors_dir = os.path.join(output_dir, "smiling_vectors")
os.makedirs(vectors_dir, exist_ok=True)

# --- 1. Save the comprehensive CSV file ---
# We are dropping the vector column from the CSV as it's not human-readable
# and we are saving the vectors separately. This makes the CSV much cleaner.
csv_path = os.path.join(output_dir, "smiling_faces_data.csv")
print(f"   - Saving summary data to: {csv_path}")
df_smiling.drop(columns=['vector']).to_csv(csv_path, index=False)

# --- 2. Save each latent vector as a separate .npy file ---
print(f"   - Saving {len(df_smiling)} individual latent vectors to: {vectors_dir}")
print("     (This is the standard format for StyleGAN analysis tools)")

for index, row in tqdm(df_smiling.iterrows(), total=df_smiling.shape[0], desc="   - Saving vectors"):
    # Get the vector data, which is currently a string
    vector_as_string = row['vector']
    seed = row['seed']

    # PARSE THE STRING BACK INTO A NUMPY ARRAY
    # 1. Clean the string by removing brackets and newlines.
    cleaned_string = vector_as_string.replace('[', '').replace(']', '').replace('\n', ' ')

    # 2. Use np.fromstring to create a flat 1D array from the space-separated numbers.
    #    This is highly efficient for this exact task.
    flat_vector = np.fromstring(cleaned_string, sep=' ')

    # 3. Reshape the vector. StyleGAN's W latent space is typically 2D (e.g., 1x512).
    #    The double brackets '[[' in the original string confirm it was a 2D array.
    #    We reshape our flat array back into the correct (1, -1) shape.
    #    Using -1 for the second dimension lets NumPy automatically infer the size (512).
    np_vector = flat_vector.reshape(1, -1).astype(np.float32)

    # Define the output filename based on the seed
    vector_filename = f"{str(seed).zfill(5)}.npy"
    vector_output_path = os.path.join(vectors_dir, vector_filename)

    # Save the now-correctly-formatted NumPy array to the .npy file
    np.save(vector_output_path, np_vector)


print("\nâœ… Step 6: All results saved successfully!")
print("You can now find your data in the 'smiling_faces_output' directory in the file browser.")
print(f"A summary CSV is at: {csv_path}")
print(f"The individual latent vectors (.npy files) are in: {vectors_dir}")
```
